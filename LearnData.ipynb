### PreparaÃ§Ã£o dos dados
1. RecÃªncia: Dias des da ultima compra atÃ© a data de referÃªncia
2. FrequÃªncia: NÃºmero total de compras realizadas
3. Valor monetÃ¡rio: Valor total gasto pelo cliente

* intervalos mÃ©dios de compras
* TencÃªncia de gastos
* Produtos mais compradas ou categoria preferida

```python


# Assuming df_adjust is your cleaned DataFrame
# Create a reference date (you can change this to any date you want)
reference_date = df_adjust['InvoiceDate'].max() + pd.Timedelta(days=1)

# Calculate Recorrencia
rfm_data = df_adjust.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})

# Rename columns
rfm_data.rename(
    columns={
        'InvoiceDate': 'Recency',
        'InvoiceNo': 'Frequency',
        'TotalPrice': 'MonetaryValue'
    },
    inplace=True
)

# Display the RFM data
rfm_data.head()

# Now you have a DataFrame 'rfm_data' containing Recency, Frequency, and MonetaryValue for each customer.
# You can proceed with further analysis like segmenting customers based on these metrics.
# Example of segmentation (you can customize the quartiles or use other methods)
quantiles = rfm_data.quantile(q=[0.25, 0.5, 0.75])
quantiles = quantiles.to_dict()

segmented_rfm = rfm_data.copy()

def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4

def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1

segmented_rfm['r_quartile'] = segmented_rfm['Recency'].apply(RScore, args=('Recency',quantiles,))
segmented_rfm['f_quartile'] = segmented_rfm['Frequency'].apply(FMScore, args=('Frequency',quantiles,))
segmented_rfm['m_quartile'] = segmented_rfm['MonetaryValue'].apply(FMScore, args=('MonetaryValue',quantiles,))
segmented_rfm['RFMScore'] = segmented_rfm.r_quartile.map(str) + segmented_rfm.f_quartile.map(str) + segmented_rfm.m_quartile.map(str)
segmented_rfm.head()
```

```python
# Assuming df_adjust is your cleaned DataFrame
# Create a reference date (you can change this to any date you want)
reference_date = df_adjust['InvoiceDate'].max() + pd.Timedelta(days=1)

# Calculate Recency, Frequency, and MonetaryValue
rfm_data = df_adjust.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (reference_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalPrice': 'sum'
})

rfm_data.rename(
    columns={
        'InvoiceDate': 'Recency',
        'InvoiceNo': 'Frequency',
        'TotalPrice': 'MonetaryValue'
    },
    inplace=True
)

# Calculate average purchase interval
# Convert 'InvoiceDate' back to datetime for calculations
# -- This is the fix to enable `.dt.days` --
df_adjust['InvoiceDate'] = pd.to_datetime(df_adjust['InvoiceDate'])
purchase_intervals = df_adjust.groupby('CustomerID')['InvoiceDate'].apply(lambda x: x.diff().dt.days.dropna())
rfm_data['AvgPurchaseInterval'] = purchase_intervals.groupby(level=0).mean()

# Calculate spending trend
rfm_data['SpendingTrend'] = df_adjust.groupby('CustomerID')['TotalPrice'].apply(lambda x: (x.iloc[-1] - x.iloc[0]) / len(x))

#Most purchased products/categories
top_products = df_adjust.groupby('CustomerID')['Description'].agg(lambda x: x.value_counts().index[0])
rfm_data['MostPurchasedProduct'] = top_products

# --- FIX: Select only numerical columns for quantile calculation ---
#Combine with RFM segmentation
numerical_features = ['Recency', 'Frequency', 'MonetaryValue', 'AvgPurchaseInterval', 'SpendingTrend']
quantiles = rfm_data[numerical_features].quantile(q=[0.25, 0.5, 0.75])
# -------------------------------------------------------------------

quantiles = quantiles.to_dict()

segmented_rfm = rfm_data.copy()

def RScore(x,p,d):
    if x <= d[p][0.25]:
        return 1
    elif x <= d[p][0.50]:
        return 2
    elif x <= d[p][0.75]:
        return 3
    else:
        return 4

def FMScore(x,p,d):
    if x <= d[p][0.25]:
        return 4
    elif x <= d[p][0.50]:
        return 3
    elif x <= d[p][0.75]:
        return 2
    else:
        return 1

segmented_rfm['r_quartile'] = segmented_rfm['Recency'].apply(RScore, args=('Recency',quantiles,))
segmented_rfm['f_quartile'] = segmented_rfm['Frequency'].apply(FMScore, args=('Frequency',quantiles,))
segmented_rfm['m_quartile'] = segmented_rfm['MonetaryValue'].apply(FMScore, args=('MonetaryValue',quantiles,))
segmented_rfm['RFMScore'] = segmented_rfm.r_quartile.map(str) + segmented_rfm.f_quartile.map(str) + segmented_rfm.m_quartile.map(str)

segmented_rfm.head()
```
